```python
#dependencies needed

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read_csv
dataset_path = "../dataset/healthcare-dataset-stroke-data.csv"
data_df = pd.read_csv(dataset_path)

print(f"The shape of the dataframe : {data_df.shape}")
print(f"Top 5 rows ")
print({data_df.head})
print(f"Column Datatype : {data_df.dtypes}")

```

```python
#Null Values

print(f" BEfore: The Null count for the columns : ")
print(f"{data_df.isnull().sum()}")

print(f" The Null percentage for the columns : ")
print(f"{((data_df.isnull().sum()/ data_df.shape[0]) * 100).round(2)}")

#Type conversion
#replaced  double quotes
data_df["bmi"] = data_df["bmi"].str.replace('"',' ', regex=True)
data_df["bmi"]=pd.to_numeric(data_df["bmi"],errors='coerce')

copy_data = data_df.copy()

data_df = data_df.fillna(data_df["bmi"].median())

print(f" After: The Null count for the columns : ")
print(f"{data_df.isnull().sum()}")
```

```python
#Duplicates
print(f"Memory Usage (Before) : {data_df.memory_usage(deep=True).sum()}")

print(f"Before Duplicate count : {data_df.duplicated().sum()}")
data_df = data_df.drop_duplicates()

print(f"After Duplicate count : {data_df.duplicated().sum()}")

print(f" The Null percentage for the columns : ")
print(f"{((data_df.isnull().sum()/ data_df.shape[0]) * 100).round(2)}")

data_df["Residence_type"] = data_df["Residence_type"].astype("category")
print(f" {data_df.dtypes}")
print(f"Memory Usage (After) : {data_df.memory_usage(deep=True).sum()}")
print(f"{data_df.isnull().sum()}")
```

```python
print(f" Describe : ")
print(f" {data_df.describe()}")

num_cols = ["id","age","hypertension","heart_disease","avg_glucose_level","bmi","stroke"]
print()
print(f"Skewness for each column:")
print()
for col in num_cols:
    print(f" {col} : { data_df[col].skew()}")
```

```python
#avg_glucose_level

q1_glucose = data_df["avg_glucose_level"].quantile(0.25)
q3_glucose = data_df["avg_glucose_level"].quantile(0.75)

IQR_Glucose = q3_glucose - q1_glucose
lower_bound = q1_glucose - 1.5 * IQR_Glucose
upper_bound = q3_glucose + 1.5 * IQR_Glucose

print(f" Quantile for Column : AVG_GLUCOSE_LEVEL ")
print("-"*10)
print(f" IQR : {IQR_Glucose}")
print(f" Lower Bound : {lower_bound}")
print(f" Upper Bound : {upper_bound}")
outliers = data_df[
    (data_df["avg_glucose_level"] < lower_bound) |
    (data_df["avg_glucose_level"] > upper_bound)
]

print("Number of outliers:", len(outliers))
print()

q1_bmi = data_df["bmi"].quantile(0.25)
q3_bmi = data_df["bmi"].quantile(0.75)

IQR_Bmi = q3_bmi - q1_bmi
lower_bound = q1_bmi - 1.5 * IQR_Bmi
upper_bound = q3_bmi + 1.5 * IQR_Bmi

print(f" Quantile for Column : BMI ")
print("-"*10)
print(f" IQR : {IQR_Bmi}")
print(f" Lower Bound : {lower_bound}")
print(f" Upper Bound : {upper_bound}")

outliers = data_df[
    (data_df["bmi"] < lower_bound) |
    (data_df["bmi"] > upper_bound)
]

print("Number of outliers:", len(outliers))
```

```python
sorted_value = data_df["bmi"].sort_values()

mean_bmi = data_df.groupby("gender")["bmi"].mean()

fig,ax = plt.subplots(1 ,2 , figsize=(12,5))
ax[0].plot(sorted_value.values)
ax[0].set_title("Sorted BMI")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("BMI")

ax[1].bar(mean_bmi.index, mean_bmi.values)
ax[1].set_title("Compare Mean BMI with Gender")
ax[1].set_xlabel("gender")
ax[1].set_ylabel("Mean BMI")

plt.tight_layout()
plt.savefig("../images/plot_and_bar_chart.png")

plt.show()
```
